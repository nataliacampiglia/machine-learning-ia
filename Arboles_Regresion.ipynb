```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

 # Datos


```python
# Datos
cars = pd.read_csv('mtcars.csv')
cars
```

```python
# Estructura y tipo de variables
cars.info()
```

```python
X = cars['wt'].values.reshape(-1,1)
y = cars['mpg'].values
```

```python
print(f'Shape X: {X.shape}')
print(f'Shape y: {y.shape}')
```

```python
# Scatterplot de datos disponibles
plt.figure(figsize=(8,3.5))
plt.scatter(X, y, label='Datos')
plt.ylabel('Consumo (mpg)')
plt.xlabel('Peso (pounds)')
plt.title('Datos disponibles - Cars Dataset')
plt.show()
```

# Arbol de profundidad cero

```python
plt.figure(figsize=(4,2.5))
plt.axhline(y=np.mean(y), label="Predicción")
plt.scatter(X, y, label="Datos")
plt.ylabel('Consumo (mpg)')
plt.xlabel('Peso (pounds)')
plt.title('Promedio de y')
plt.legend()
plt.show()
```

# Arbol de profundidad 1

```python
v_range = np.linspace(1,6,1000)

RSS = []

for v in v_range:
    c1 = cars.loc[cars['wt']<=v,'mpg'].mean()
    c2 = cars.loc[cars['wt']>v,'mpg'].mean()
    rss = ((cars.loc[cars['wt']<=v,'mpg']-c1)**2).sum()+((cars.loc[cars['wt']>v,'mpg']-c2)**2).sum()
    RSS.append(rss)
```

```python
plt.figure(figsize=(4,4.5))
plt.step(v_range,np.array(RSS)/len(cars), label='MSE')
plt.axvline(x=((2.200+2.320)/2),linestyle='dashed',color='black', label='Mejor v')
plt.xlabel('Valor de v')
plt.ylabel('MSE')
plt.legend()
plt.title(r'¿Cuál es el mejor $v$?')
plt.show()
```

```python
def h(x):
    if x <= (2.200+2.320)/2:
        return cars.loc[cars['wt']<=(2.200+2.320)/2,'mpg'].mean()
    else:
        return cars.loc[cars['wt']>(2.200+2.320)/2,'mpg'].mean()
```

```python
x_range = np.linspace(1,6,1000)
vec_h = [h(x) for x in x_range]
plt.figure(figsize=(4,4.5))
plt.step(x_range, vec_h, label="Predicción")
plt.scatter(X, y, label="Datos")
plt.axvline(x=((2.200+2.320)/2),linestyle='dashed',color='black', label = 'v')
plt.legend()
plt.ylabel('Consumo (mpg)')
plt.xlabel('Peso (pounds)')
plt.title(r'La mejor $h$ con una pregunta')
plt.show()
```

# Arbol de cualquier profundidad

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Elegimos la profundidad
depth=None
```

```python
# Instanciamos el arbol
tree = DecisionTreeRegressor(max_depth=depth)
```

```python
# ERM con .fit (entrenamos) arbol entrenado
tree.fit(X, y)
```

```python
# Hacemos predicciones con .predict
# Este es el modelo
y_pred = tree.predict(X)
```

```python
# Root mean squared error
np.sqrt(mean_squared_error(y,y_pred))
```

```python
#### Graficamos la hipótesis obtenida
X_ax = np.linspace(1,6,100).reshape(-1, 1)
y_pred = tree.predict(X_ax)

plt.figure(figsize=(4,4.5))
plt.scatter(X, y, color="black", label="Datos")
plt.step(X_ax, y_pred, lw=2.5, label="Predicción")
plt.ylabel('Consumo (mpg)')
plt.xlabel('Peso (pounds)')
plt.title("Profundidad "+str(depth))
plt.legend()
plt.show()
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(20,20))
plot_tree(tree, filled=True, feature_names=['Peso'])
plt.show()
```
